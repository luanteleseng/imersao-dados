## Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
```

## Loading Data

```python
dados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true', 
                    compression = 'zip')
dados.rename(columns = {'droga' : 'composto'}, inplace = True)

dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados.head()
```

```python
#mecanismo ativado (1), não ativado (0)
dados_resultados['acat_inhibitor'].unique()
```

```python
dados_resultados['n-moa'] != 0
```

```python
dados_resultados['n-moa'] = dados_resultados.drop('id', axis = 1).sum(axis = 1)
```

```python
dados_resultados['ativo-moa'] = (dados_resultados['n-moa'] != 0)
dados_resultados.head()
```

```python
dados_combinados = pd.merge(dados, dados_resultados[['id', 'n-moa', 'ativo-moa']], on = 'id')
dados_combinados.head()
```

## Model Training

```python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo-moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 450)

modelo_rlogistica = LogisticRegression(max_iter = 1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier(strategy = 'most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)

accuracy_score(y_teste, previsao_dummy)
```

```python
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo-moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 450)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(20, 8), facecolor = 'c')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names = x_treino.columns,
               class_names = ['Não Ativado', 'Ativado'])

plt.show()
```

```python
x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo-moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 376)

teste = []
treino = []
for i in range(1, 15):
    modelo_arvore = DecisionTreeClassifier(max_depth = i)
    modelo_arvore.fit(x_treino, y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```
