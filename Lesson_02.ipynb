## Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

## Loading Data

```python
# Carregando o Dataframe e visualizando
url_data = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
df = pd.read_csv(url_data, compression = 'zip')
df.head() # dataframe visualization
```

```python
# Substituindo a string '-' nas colunas
df.columns = [str(i).replace('-','') for i in df.columns.to_list()]
df.rename(columns = {'droga' : 'composto'}, inplace = True) # inplace True para modificar no dataframe
df.head()
```

## Plotting

```python
compost = df['composto'].value_counts().index[:5]
compost
```

```python
df.query('composto in @compost')
```

```python
sns.set() # estilização
# plt.figure(figsize = (7,5))
ax = sns.countplot(x = 'composto', data = df.query('composto in @compost'))
ax.set_title('largest five compounds')
plt.show() # apresenta o gráfico sem algumas infos
```
